```python
#We are given a function F and we define a function f(K) - given function - F
#We want to find the root of that function using Newton's method, which will
#then be solution for K in original function
def func(K):
    return (L/K)*(1-np.exp(-K*(t+a)))-F
```

```python
#We will need the derivative of that function for Newton's method
def func_deriv(K):
    return (-L/K**2)*(1-np.exp(-K*(t+a)))+(L/K)*(t+a)*np.exp(-K*(t+a))
```

```python
#Let's plot the function we want to be working with to see if it smooth enough
#that an application of Newton's method is appropriate
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
L = 1
t = 1
a = 1
F = 1

K = np.linspace(0, 10, 1000)

plt.plot(K,func(K))
```

```python
#Newton's method is an iterative scheme defined as x_{n+1} = x_n - f(x_)
def newtons(Tol, x_0 = 0):
    x_old = x_0
    x_new = x_old + Tol
    while np.abs(x_new-x_old) >= Tol:
        (x_old, x_new) = (x_new, x_new - func(x_new)/func_deriv(x_new))
    print('The estimate for K is {:f}.'.format(x_new))
```
