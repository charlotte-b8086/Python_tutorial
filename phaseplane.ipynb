Phase plane vector field

```python
import numpy as np
import matplotlib.pyplot as plt

S = np.linspace(0,1,31)
I = np.linspace(0,1,31)

X,Y = np.meshgrid(S,I)

beta = 0.5
nu = 0.3

dSdt = -beta*X*Y
dIdt = beta*X*Y - nu*Y
#dRdt = nu*Y
```

```python
plt.quiver(X, Y, dSdt, dIdt, pivot='mid', angles='xy')
plt.axis('square')
plt.xlabel('S')
plt.ylabel('I')
```

```python
plt.streamplot(X, Y, dSdt, dIdt)
plt.xlabel('S')
plt.ylabel('I')
plt.plot(S,1-I)
plt.xlim(0,1)
```
